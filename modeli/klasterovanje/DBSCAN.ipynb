# DBSCAN

```python
import pandas as pd
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import MinMaxScaler 
from matplotlib import pyplot as plt
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import dendrogram, linkage
```

Ucitavanje podataka

```python
data = pd.read_csv('/home/user/Desktop/IP/sredjeni_podaci.csv')
print(data.shape)
```

```python
data.drop('Unnamed: 0',axis=1,inplace= True)
```

Brisemo ciljnu promenljivu,jer je ovo nenadgledno ucenje.

```python
X = data.drop('match',axis=1)
```

Standardizacija podataka:

```python
feature_names = X.columns
scaler = MinMaxScaler()
X = pd.DataFrame(scaler.fit_transform(X), columns=feature_names)
```

```python
model = DBSCAN(eps=0.1, min_samples=2)
```

```python
model.fit(X)
```

```python
print(np.unique((model.labels_)))
```

Ovakvim izbrom eps i min_samples dobijamo da su sve instance sum.

```python
model = DBSCAN(eps=0.1, min_samples=20)
```

```python
model.fit(X)
print(np.unique((model.labels_)))
```

```python
model = DBSCAN(eps=0.5, min_samples=5)
model.fit(X)
print(np.unique((model.labels_)))
```

Zbog velikog broja instanci ne mogu da vizualizujem klase i da tako vidim koji su najbolji parametri ovog modela, ali cu probati da na osnovu vrednosti silueta skora to vidim.

```python
min_samples_value = range(2,50);
eps_values = [0.1, 0.15, 0.2, 0.25, 0.3]
X = data.drop('match',axis=1)
for eps in eps_values:
    for min_samples in min_samples_value:
        model = DBSCAN(eps=eps, min_samples=min_samples)
        model.fit(X);
        print("Jedinstvene vrednsoti klasa za eps = ",eps," i min samples = ",min_samples, "su: ", np.unique(model.labels_))
        if len(np.unique(model.labels_)) == 1:
            s = None
        else:
            s = silhouette_score(X, model.labels_)
        print("Silueta skor za eps = ",eps," i min samples = ",min_samples, "je : ",s);
```

Svaki put dobijam da su sve instance sumovi,algoritam ne pronalazi nijednu klasu.
