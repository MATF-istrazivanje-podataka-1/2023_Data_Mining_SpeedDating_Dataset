Nas ciljni atrbut ima samo dve klase (0 i 1), pa zakljucujemo da se radi o binarnoj klasifikaciji

Zelimo da vidimo kako su raspodeljene klase, odnosno koliko ima istanci prve, a koliko istanci druge klase 

```python
br1 = 0;
br0 = 0;
for y in Y:
    if y == 1:
        br1+=1
    else:
        br0+=1
print('Broj instanci prve klase: ', br0)
print('Broj instanci druge klase: ',br1)
```

Primecujemo da je broj instanci prve klase znatno veci od broja instanci druge klase.

Sada podatke delimo na TRENING i TEST skup. 75% ce biti podaci za trening, dok su 25% podaci za testiranje. stratify koristimo da se ne bi desilo da u trening skupu imamo samo instance koje pripadaju prvoj klasi, vec da imamo podatke iz obe klase

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, stratify=Y, random_state=12)
```
